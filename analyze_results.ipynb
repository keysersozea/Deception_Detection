```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Adam_lr0001_WDe_2 = pd.read_csv('Results/Adam_lr0001-WDe-2.csv')
Adam_lr0001_WDe_2["Title"] = "Adam & LR 0.0001 % & WD 10e-2"

Adam_lr001_WDe_2 = pd.read_csv('Results/Adam_lr001-WDe-2.csv')
Adam_lr001_WDe_2["Title"] = "Adam & LR 0.001 % & WD 10e-2"

Adam_lr0001Schedular_WDe_2 = pd.read_csv('Results/Adam_lr0001Schedular-WDe-2.csv')
Adam_lr0001Schedular_WDe_2["Title"] = "Adam & LR 0.0001 % & WD 10e-2 & LR Schedular"

Adam_lr001Schedular_WDe_2 = pd.read_csv('Results/Adam_lr001Schedular-WDe-2.csv')
Adam_lr001Schedular_WDe_2["Title"] = "Adam & LR 0.001 % & WD 10e-2 & LR Schedular"

SGD_lr0001Schedular_WDe_2 = pd.read_csv('Results/SGD_lr0001Schedular-WDe-2.csv')
SGD_lr0001Schedular_WDe_2["Title"] = "SGD & LR 0.0001 % & WD 10e-2 & LR Schedular"

SGD_lr001Schedular_WDe_2 = pd.read_csv('Results/SGD_lr001Schedular-WDe-2.csv')
SGD_lr001Schedular_WDe_2["Title"] = "SGD & LR 0.001 % & WD 10e-2 & LR Schedular"

results = [Adam_lr0001_WDe_2, Adam_lr001_WDe_2, Adam_lr0001Schedular_WDe_2, Adam_lr001Schedular_WDe_2, SGD_lr001Schedular_WDe_2, SGD_lr0001Schedular_WDe_2]
```

```python
plt.figure(figsize=(25, 20))   
for i, data in enumerate(results):
    plt.subplot(2, 3, i+1)  # 2 rows, 3 columns, index i+1
    plt.plot(data['epoch']+1, data['trainloss'], label='Train Loss')
    plt.plot(data['testloss'], label='Test Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title(data['Title'][0])
    plt.legend()
```
